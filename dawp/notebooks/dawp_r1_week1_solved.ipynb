Î Î±ÏÎ±Ï„Î®ÏÎ·ÏƒÎ·: Î— Î±Ï€ÏŒÏƒÏ„Î±ÏƒÎ· Î¼ÎµÏ„Î±Î¾Ï Ï„Î¿Ï… ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… (x1, y1) ÎºÎ±Î¹ Ï„Î¿Ï… ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… (x2, y2) ÏƒÏ„Î¿ ÎºÎ±ÏÏ„ÎµÏƒÎ¹Î±Î½ÏŒ ÎµÏ€Î¯Ï€ÎµÎ´Î¿ Î´Î¯Î½ÎµÏ„Î±Î¹ Î±Ï€ÏŒ Ï„Î¿Î½ Ï„ÏÏ€Î¿: $\sqrt{(x_2 - x_1)^2 + (y_2-y_1)^2}$

````python
# Î›ÏÏƒÎ· 1.3
import math

def distance(x1, y1, x2=0, y2=0):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

xA = float(input("Î”ÏÏƒÎµ x ÏƒÏ…Î½Ï„ÎµÏ„Î±Î³Î¼Î­Î½Î· ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… A: "))
yA = float(input("Î”ÏÏƒÎµ y ÏƒÏ…Î½Ï„ÎµÏ„Î±Î³Î¼Î­Î½Î· ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… A: "))
xB = float(input("Î”ÏÏƒÎµ x ÏƒÏ…Î½Ï„ÎµÏ„Î±Î³Î¼Î­Î½Î· ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… B: "))
yB = float(input("Î”ÏÏƒÎµ y ÏƒÏ…Î½Ï„ÎµÏ„Î±Î³Î¼Î­Î½Î· ÏƒÎ·Î¼ÎµÎ¯Î¿Ï… B: "))

dist_AB = distance(xA, yA, xB, yB)

dist_A_origin = distance(xA, yA)
dist_B_origin = distance(xB, yB)

print(f"Î‘Ï€ÏŒÏƒÏ„Î±ÏƒÎ· Î¼ÎµÏ„Î±Î¾Ï A({xA}, {yA}) ÎºÎ±Î¹ B({xB}, {yB}): {dist_AB:.2f}")
print(f"Î‘Ï€ÏŒÏƒÏ„Î±ÏƒÎ· A({xA}, {yA}) Î±Ï€ÏŒ Ï„Î·Î½ Î±ÏÏ‡Î® (0,0): {dist_A_origin:.2f}")
print(f"Î‘Ï€ÏŒÏƒÏ„Î±ÏƒÎ· B({xB}, {yB}) Î±Ï€ÏŒ Ï„Î·Î½ Î±ÏÏ‡Î® (0,0): {dist_B_origin:.2f}")
````

````python
# Î›ÏÏƒÎ· 1.4
import codecs
import this

# Î‘Ï€Î¿ÎºÏ‰Î´Î¹ÎºÎ¿Ï€Î¿Î¯Î·ÏƒÎ· ÎºÎµÎ¹Î¼Î­Î½Î¿Ï… Î¼Îµ ROT13
s = codecs.encode(this.s, "rot13")

# 1. ÎšÎ±Î¸Î±ÏÎ¹ÏƒÎ¼ÏŒÏ‚ ÎºÎµÎ¹Î¼Î­Î½Î¿Ï…
clean_text = (
    s.replace("\n", " ")
    .replace("*", "")
    .replace("-", "")
    .replace(",", "")
    .replace(".", "")
    .upper()
)

# 2. Î›Î¯ÏƒÏ„Î± Î»Î­Î¾ÎµÏ‰Î½ Ï‡Ï‰ÏÎ¯Ï‚ Î´Î¹Ï€Î»ÏŒÏ„Ï…Ï€Î±
words = clean_text.split()
unique_words = list(set(words))

# 3. Î›ÎµÎ¾Î¹ÎºÏŒ Î¼Îµ Ï€Î»Î®Î¸Î¿Ï‚ ÎµÎ¼Ï†Î¬Î½Î¹ÏƒÎ·Ï‚ ÎºÎ¬Î¸Îµ Î»Î­Î¾Î·Ï‚
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print("\nÎ›Î­Î¾ÎµÎ¹Ï‚ Ï„Î±Î¾Î¹Î½Î¿Î¼Î·Î¼Î­Î½ÎµÏ‚ Î±Î»Ï†Î±Î²Î·Ï„Î¹ÎºÎ¬ Î¼Îµ Ï„Î· ÏƒÏ…Ï‡Î½ÏŒÏ„Î·Ï„Î¬ Ï„Î¿Ï…Ï‚:\n")
for w in sorted(word_count.keys()):
    print(f"{w}: {word_count[w]}")

# 4. Î–Î®Ï„Î·ÏƒÎ· Î»Î­Î¾Î·Ï‚ Î±Ï€ÏŒ Ï‡ÏÎ®ÏƒÏ„Î· ÎºÎ±Î¹ ÎµÎ¼Ï†Î¬Î½Î¹ÏƒÎ· Î±Ï€Î¿Ï„ÎµÎ»Î­ÏƒÎ¼Î±Ï„Î¿Ï‚
user_word = input("Î”ÏÏƒÎµ Î¼Î¹Î± Î»Î­Î¾Î·: ").upper()
if user_word in word_count:
    print(
        f"Î— Î»Î­Î¾Î· '{user_word}' ÎµÎ¼Ï†Î±Î½Î¯Î¶ÎµÏ„Î±Î¹ {word_count[user_word]} Ï†Î¿ÏÎ­Ï‚ ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿."
    )
else:
    print(f"Î— Î»Î­Î¾Î· {user_word} Î´ÎµÎ½ ÎµÎ¯Î½Î±Î¹ Ï€Î¿Î»Ï Î¶ÎµÎ½!")
````

````python
# Î›ÏÏƒÎ· 1.5

from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles. Determines return value units.
    return c * r


athens = (23.7348324, 37.9755648) # (longitude, latitude)
berlin = (13.3989367, 52.510885) # (longitude, latitude)

print(haversine(*athens, *berlin))

from haversine import haversine, Unit

print(haversine(athens[::-1], berlin[::-1], unit=Unit.KILOMETERS))
````
