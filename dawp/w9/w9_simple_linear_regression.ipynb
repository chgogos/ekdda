# Απλή γραμμική παλινδρόμηση

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn   
```

```python
# Δημιουργία συνθετικών δεδομένων
np.random.seed(42)
n = 20
x = np.linspace(0, 10, n)
y = 3 * x  + 2 + np.random.randn(n)
print(x)
print(y)
plt.scatter(x, y);
```

```python
# Το x ως διάνυσμα στήλης (χρειάζεται να είναι σε αυτή 
# τη μορφή για να μπορεί να χρησιμοποιηθεί στο model.fit)
x[:, np.newaxis]
x.reshape(-1, 1)  # 2ος εναλλακτικός τρόπος
np.vstack([x]).T  # 3ος εναλλακτικός τρόπος
```

```python
# Προσαρμογή γραμμής στα σημεία (x,y)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x[:, np.newaxis], y)
xfit = np.linspace(0, 10, 100)
yfit = model.predict(xfit[:, np.newaxis])
```

```python
# Σχεδίαση σημείων και γραμμής παλινδρόμησης
plt.plot(xfit, yfit, color="black")
plt.plot(x, y, "o");
```

```python
# Εκτύπωση συντελεστών της γραμμής παλινδρόμησης
print("Παράμετροι:", model.coef_, model.intercept_)
slope = model.coef_[0]
intercept = model.intercept_
print(f"y={slope:.2f}x + {intercept:.2f}")  
```
