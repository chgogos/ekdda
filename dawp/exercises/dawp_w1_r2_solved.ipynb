1.1 Να γράψετε κώδικα που να υπολογίζει το συνολικό κόστος κατανάλωσης νερού ενός νοικοκυριού, σύμφωνα με την ακόλουθη **κλιμακωτή χρέωση**:

| Κατανάλωση ($m^3$) | Τιμή ανά $m^3$ (σε ευρώ) |
|------------------|----------------|
| 0 έως και 30     | 0.50           |
| >30 έως και 50   | 0.75           |
| >50 έως και 80   | 1.00           |
| >80              | 1.50           |

Για παράδειγμα η χρέωση για κατανάλωση 55 $m^3$ νερού είναι: 30 * 0.5 + 20 * 0.75 + 5 * 1 = 35 ευρώ.

Ο κώδικας θα πρέπει να:

1. Διαβάζει από τον χρήστη την κατανάλωση νερού (σε κυβικά μέτρα).  
2. Υπολογίζει το συνολικό κόστος με βάση την κλιμακωτή χρέωση.  
3. Εμφανίζει το αποτέλεσμα με δύο δεκαδικά ψηφία.

````python
# Λύση 1.1
consumption = float(input("Δώσε την κατανάλωση νερού σε m^3: "))

cost = 0.0

if consumption <= 30:
    cost = consumption * 0.50
elif consumption <= 50:
    cost = 30 * 0.50 + (consumption - 30) * 0.75
elif consumption <= 80:
    cost = 30 * 0.50 + 20 * 0.75 + (consumption - 50) * 1.00
else:
    cost = 30 * 0.50 + 20 * 0.75 + 30 * 1.00 + (consumption - 80) * 1.50

# Εμφάνιση αποτελέσματος
print(f"Συνολικό κόστος: {cost:.2f} ευρώ")
````

1.2 Γράψτε κώδικα που να καταμετρά για κάθε τετραψήφιο θετικό ακέραιο αριθμό `x` το πλήθος των ψηφίων του που διαιρεί ακριβώς τον ίδιο τον αριθμό (για παράδειγμα ο αριθμός 1234 διαιρείται με τους y, συνεπώς το πλήθος των ψηφίων του που τον διαιρούν είναι z). Ο κώδικας να εμφανίζει τη μέση τιμή του πλήθους των ψηφίων για όλους τους τετραψήφιους θετικούς αριθμούς.

````python
# Λύση 1.2
count = 0      
total = 0    

for x in range(1000, 10000):
    div_count = 0
    s = str(x)
    for ch in s:
        d = int(ch)
        if d != 0 and x % d == 0:
            div_count += 1
    count += div_count
    total += 1

average = count / total
print(f"Μέση τιμή πλήθους ψηφίων που διαιρούν τους τετραψήφιους αριθμούς: {average}")
````

1.3 Γράψτε μια συνάρτηση `is_disarium(x)` που να ελέγχει αν ένας αριθμός είναι disarium αριθμός. Διαβάστε τι είναι disarium αριθμός στο <https://rosettacode.org/wiki/Disarium_numbers>. Εκτυπώστε όλους τους disarium αριθμούς στο διάστημα [10, 3000].

 extra challenge: να δείχνετε για κάθε αριθμό που είναι disarium, γιατί αυτό ισχύει (π.χ. 1^1 + 3^2 + 5^3 == 1 + 9 + 125 == 135)

````python
# Λύση 1.3
def is_disarium(x):
    s = str(x)
    total = 0
    for i in range(len(s)):
        digit = int(s[i])
        total += digit ** (i + 1)
    return total == x

for n in range(10, 3001):
    if is_disarium(n):
        print(n)

````

````python
# Λύση 1.3 (extra)

def is_disarium(x):
    s = str(x)
    powers = []
    total = 0
    for i in range(len(s)):
        d = int(s[i])
        val = d ** (i + 1)
        powers.append(f"{d}^{i+1}")
        total += val
    if total == x:
        terms = " + ".join(powers)
        values = " + ".join(str(int(s[i]) ** (i + 1)) for i in range(len(s)))
        print(f"{terms} == {values} == {x}")
        return True
    return False

for n in range(10, 3001):
    is_disarium(n)
````

1.5 Να υλοποιήσετε τις ακόλουθες δύο συναρτήσεις:
- Τη συνάρτηση `hamming_distance(s1, s2)` που λαμβάνει δύο συμβολοσειρές s1 και s2 ίδιου μήκους. και επιστρέφει την απόσταση Hamming (πλήθος θέσεων όπου διαφέρουν οι χαρακτήρες στις s1 και s2, δείτε το <https://en.wikipedia.org/wiki/Hamming_distance>). Αν οι συμβολοσειρές έχουν διαφορετικό μήκος, τότε η συνάρτηση να επιστρέφει -1.
- Τη συνάρτηση `check_password_similarity(passwords)` που λαμβάνει μια λίστα passwords και για κάθε δυνατό ζεύγος passwords της λίστας, υπολογίζει την απόσταση Hamming χρησιμοποιώντας τη συνάρτηση `hamming_distance`. Η συνάρτηση να εκτυπώνει για κάθε ζεύγος, την απόσταση  Hamming  και αν θεωρούνται παρόμοια (δλδ. η απόσταση Hamming distance είναι το πολύ 2).

Καλέστε τη συνάρτηση `check_password_similarity(passwords)` ως εξής:

```
passwords = [
    "Abc123$!",    # 8 chars
    "Abd123$#",    # 8 chars
    "Xyz987@!",    # 8 chars
    "Abc124$!",    # 8 chars
    "password1",   # 9 chars
    "Password2!",  # 10 chars
    "AbC1!xY2",    # 8 chars
    "12345678",    # 8 chars
]
check_password_similarity(passwords)
```

````python
# Λύση 1.5
def hamming_distance(s1, s2):
    if len(s1) != len(s2):
        return -1
    dist = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            dist += 1
    return dist


def check_password_similarity(passwords):
    n = len(passwords)
    for i in range(n):
        for j in range(i + 1, n):
            pwd1 = passwords[i]
            pwd2 = passwords[j]
            dist = hamming_distance(pwd1, pwd2)
            if dist == -1:
                print(f"{pwd1} - {pwd2}: διαφορετικά μήκη, δεν υπολογίζεται Hamming distance")
            else:
                similar = "Παρόμοια" if dist <= 2 else "Όχι παρόμοια"
                print(f"{pwd1} - {pwd2}: Hamming distance = {dist}, {similar}")


passwords = [
    "Abc123$!",    # 8 chars
    "Abd123$#",    # 8 chars
    "Xyz987@!",    # 8 chars
    "Abc124$!",    # 8 chars
    "password1",   # 9 chars
    "Password2!",  # 10 chars
    "AbC1!xY2",    # 8 chars
    "12345678",    # 8 chars
]
check_password_similarity(passwords)
````
