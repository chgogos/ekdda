1.1 Γράψτε μια κλάση με όνομα `Player` που να αναπαριστά έναν παίκτη σε ένα βιντεοπαιχνίδι. Η κλάση να έχει ως χαρακτηριστικά όνομα (`name`) και ζωές (`lives`), καθώς και μία μέθοδο με όνομα `die()` που να μειώνει κατά μία τις ζωές του παίκτη κάθε φορά που καλέιται. Δημιουργήστε ένα αντικείμενο της κλάσης `Player`, με όνομα `"Luigi"` και 3 ζωές. Καλέστε 2 φορές τη μέθοδο `die()` και εμφανίστε το αντικείμενο. Προσθέσετε κατάλληλες μεθόδους `__init__` και `__str__`.

````python
# Λύση 1.1
class Player:
    def __init__(self, name, lives):
        self.name = name
        self.lives = lives

    def die(self):
        if self.lives > 0:
            self.lives -= 1

    def __str__(self):
        return f"Παίκτης='{self.name}', ζωές={self.lives}"


# Δημιουργία αντικειμένου
p = Player("Luigi", 3)

# Ο παίκτης χάνει δύο ζωές
p.die()
p.die()

# Εμφάνιση του αντικειμένου
print(p)
````

Αν η γραμμή δεν είναι έγκυρη, η συνάρτηση να επιστρέφει None. Στη συνέχεια, για την ακόλουθη λίστα τέτοιων γραμμών, μετρήστε πόσες έγκυρες γραμμές με WARN έχει o κάθε χρήστης.

To regex μοτίβο (pattern) που μπορείτε να χρησιμοποιήσετε για έλεγχο εγκυρότητας γραμμής και την εξαγωγή των επιμέρους στοιχείων κάθε γραμμής είναι το: 
- `r"^\[(\d{4}-\d{2}-\d{2})\] \[(INFO|WARN|ERROR)\] \[(u\d{4})\]: (.+)$"`

ενώ η εξαγωγή των επιμέρους στοιχείων της γραμμής γίνεται με:

- `match.group(1)` για date
- `match.group(2)` για level
- `match.group(3)` για user_id
- `match.group(4)` για message

````python
# Λύση 1.5
import re

def parse_log_line(line):
    pattern = r"^\[(\d{4}-\d{2}-\d{2})\] \[(INFO|WARN|ERROR)\] \[(u\d{4})\]: (.+)$"
    match = re.match(pattern, line)
    if match:
        return {
            "date": match.group(1),
            "level": match.group(2),
            "user_id": match.group(3),
            "message": match.group(4)
        }
    return None


log_lines = [
    "[2025-10-21] [INFO] [u1234]: User logged in",
    "[2025-10-21] [WARN] [u5678]: Password attempt failed",
    "[2025-10-20] [ERROR] [u9876]: Connection timeout",
    "[2025-10-19] [INFO] [u1234]: File uploaded",
    "[2025-10-19] [ERROR] [u5678]: Disk full",
    "[2025-10-18] [WARN] [u4321]: Session expired",
    "[2025-10-18] [INFO] [u8765]: User logged out",
    "[2025-10-17] [ERROR] [u1234]: Failed to save document",
    "[2025-10-17] [INFO] [u5678]: Password changed",
    "[2025-10-16] [WARN] [u9876]: Low memory warning",
    "[2025-10-16] [ERROR] [u4321]: Unauthorized access attempt",
    "[2025-10-15] [INFO] [u8765]: Profile updated",
    "[2025-10-15] [WARN] [u1234]: Too many login attempts",
    "[2025-10-14] [ERROR] [u5678]: Database connection lost",
    "[2025-10-14] [INFO] [u9876]: User registered",
    "[2025-10-13] [WARN] [u4321]: Password expiry soon",
    "[2025-10-13] [ERROR] [u8765]: Email sending failed",
    "[2025-10-12] [INFO] [u1234]: Download completed",
    "[2025-10-12] [WARN] [u5678]: High CPU usage",
    "[2025-10-11] [ERROR] [u9876]: Server unreachable",
    # Λανθασμένες γραμμές
    "[2025/10/21] [INFO] [u1234]: User logged in",
    "[2025-10-21] [DEBUG] [u5678]: Debugging message",
    "[2025-10-20] [ERROR] [1234]: Missing 'u' στο user_id",
    "[2025-10-19] [INFO] [u12a4]: user_id δεν έχει 4 ψηφία",
]

# Μετράμε τις έγκυρες γραμμές με WARN ανά χρήστη
warn_count = {}

for line in log_lines:
    result = parse_log_line(line)
    if result and result["level"] == "WARN":
        user = result["user_id"]
        if user not in warn_count:
            warn_count[user] = 0
        warn_count[user] += 1

# Εκτύπωση αποτελεσμάτων
print("WARN counts per user:")
for user, count in warn_count.items():
    print(f"{user}: {count}")
````
