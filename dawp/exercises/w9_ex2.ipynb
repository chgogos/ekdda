Γραμμική παλινδρόμηση με 2 ερμηνευτικές μεταβλητές

```python
import numpy as np
from sklearn.linear_model import LinearRegression


X = [[0, 0], [1, 1], [2, 3], [3, 2]]
y = [0.0, 5.00, 13.50, 10.90]

model = LinearRegression()
model.fit(X, y)

yfit = model.predict([[4, 2]])
print(yfit)
```
