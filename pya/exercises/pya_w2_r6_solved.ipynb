### Παράμετροι και ορίσματα
2.1 Γράψτε μια συνάρτηση που να δέχεται δύο λεκτικά και να επιστρέφει την απόσταση Hamming (δηλαδή το πλήθος των αντίστοιχων χαρακτήρων που διαφέρουν στα δύο λεκτικά). Αν τα λεκτικά είναι διαφορετικού μήκους η συνάρτηση να επιστρέφει την τιμή -1. Για τα ακόλουθα λεκτικά: HOUSE, MOUSE, HORSE, WORSE, NURSE, VERSE, CAUSE, PAUSE, CASES, CARES, βρείτε τις αποστάσεις hamming για κάθε πιθανό ζεύγος λεκτικών.

```python
# Λύση 2.1

def hamming_distance(s, t):
    if len(s) != len(t):
        return -1
    c = 0
    for i in range(len(s)):
        if s[i] != t[i]:
            c += 1
    return c


words = [
    "HOUSE", "MOUSE", "HORSE", "WORSE", "NURSE",
    "VERSE", "CAUSE", "PAUSE", "CASES", "CARES"
]

n = len(words)

for i in range(n):
    for j in range(i + 1, n):
        d = hamming_distance(words[i], words[j])
        print(f"d({words[i]}, {words[j]}) = {d}")
```

```python
# Λύση 2.1 (pythonic)
from itertools import combinations

def hamming_distance(s1, s2):
    if len(s1) != len(s2):
        return -1
    return sum(c1 != c2 for c1, c2 in zip(s1, s2))


words = [
    "HOUSE", "MOUSE", "HORSE", "WORSE", "NURSE",
    "VERSE", "CAUSE", "PAUSE", "CASES", "CARES"
]

for w1, w2 in combinations(words, 2):
    d = hamming_distance(w1, w2)
    print(f"d({w1}, {w2}) = {d}")
```

## Προαιρετικά ορίσματα, προκαθορισμένες τιμές συναρτήσεων
2.2 Γράψτε μια συνάρτηση count_char(text, char) που να επιστρέφει πόσες φορές υπάρχει ο χαρακτήρας char στο κείμενο text. Το όρισμα char να είναι προαιρετικό και αν παραληφθεί να χρησιμοποιείται ο χαρακτήρας ενός διαστήματος ' '. Χρησιμοποιώντας τη συνάρτηση εμφανίστε πόσες φορές υπάρχει κάθε κεφαλαίο γράμμα του αγγλικού αλφαβήτου στην πρόταση "I TRIED TO FIX A BUG BUT ACCIDENTALLY TURNED IT INTO A FEATURE". Μπορείτε να λάβετε τα γράμματα του αγγλικού αλφαβήτου από το module string της Python, χρησιμοποιώντας τη σταθερά string.ascii_uppercase. 

```python
# Λύση 2.2
import string

def count_char(text, char=' '):
    return text.count(char)

sentence = "I TRIED TO FIX A BUG BUT ACCIDENTALLY TURNED IT INTO A FEATURE"

for letter in string.ascii_uppercase:
    occurrences = count_char(sentence, letter)
    print(f"{letter}: {occurrences}")
```

## Ορίσματα τύπου *args

2.3 Σε ένα πειραματικό σύστημα τηλεμετρίας, κάθε αισθητήρας στέλνει μια ακολουθία από τιμές άγνωστου πλήθους κάθε φορά που ενεργοποιείται. Να υλοποιήσετε συνάρτηση telemetry_summary(*args) η οποία δέχεται ως ορίσματα έναν αυθαίρετο αριθμό αριθμητικών τιμών (ακέραιων ή πραγματικών), καθεμία από τις οποίες αντιστοιχεί σε μία μέτρηση αισθητήρα. Η συνάρτηση να επιστρέφει τα εξής στοιχεία: (α) το πλήθος των μετρήσεων, (β) τη μέση τιμή τους, (γ) τη μέγιστη και (δ) την ελάχιστη τιμή. Αν η συνάρτηση κληθεί χωρίς ορίσματα, να επιστρέφει None. Δοκιμάσετε τη συνάρτηση με διαφορετικό αριθμό μετρήσεων και να εμφανίσετε τα αποτελέσματα με κατάλληλα μηνύματα.

```python
# Λύση 2.3
def telemetry_summary(*args):
    if not args:
        return None

    count = len(args)
    average = sum(args) / count
    maximum = max(args)
    minimum = min(args)

    return count, average, maximum, minimum


# Δοκιμές της συνάρτησης
measurements1 = telemetry_summary(10, 12.5, 9, 15, 11)
measurements2 = telemetry_summary(3.2, 7.8)
measurements3 = telemetry_summary()

if measurements1 is not None:
    print(f"Μετρήσεις 1 -> πλήθος: {measurements1[0]}, μέση τιμή: {measurements1[1]:.2f}, μέγιστη: {measurements1[2]}, ελάχιστη: {measurements1[3]}")

if measurements2 is not None:
    print(f"Μετρήσεις 2 -> πλήθος: {measurements2[0]}, μέση τιμή: {measurements2[1]:.2f}, μέγιστη: {measurements2[2]}, ελάχιστη: {measurements2[3]}")

print(f"Μετρήσεις 3 -> αποτέλεσμα: {measurements3}")
```

## Εμβέλεια μεταβλητών
2.4 Σε ένα απλοποιημένο σύστημα καταγραφής θερμοκρασίας, υπάρχει μια μεταβλητή threshold που δηλώνει το όριο ασφαλείας της θερμοκρασίας και είναι ορισμένη σε καθολικό (global) επίπεδο με αρχική τιμή 25. Να υλοποιήσετε συνάρτηση check_temperature(temp) η οποία δέχεται ως όρισμα μια τιμή θερμοκρασίας και χρησιμοποιώντας μόνο ανάγνωση της global μεταβλητής threshold εμφανίζει το μήνυμα "SAFE" όταν η θερμοκρασία είναι μικρότερη ή ίση του ορίου και "WARNING" όταν το υπερβαίνει. Στη συνέχεια, να υλοποιήσετε συνάρτηση update_threshold(new_value) η οποία τροποποιεί τη global μεταβλητή threshold ώστε να λαμβάνει νέα τιμή. Στο κυρίως πρόγραμμα να καλέσετε τις συναρτήσεις πριν και μετά την αλλαγή του ορίου.

```python
# Λύση 2.4
threshold = 25

def check_temperature(temp):
    if temp <= threshold:
        print("SAFE")
    else:
        print("WARNING")

def update_threshold(new_value):
    global threshold
    threshold = new_value


# Κυρίως πρόγραμμα
check_temperature(23)
check_temperature(30)

update_threshold(22)

check_temperature(23)
check_temperature(30)
```
