1.2 Γράψτε πρόγραμμα που να ζητά από τον χρήστη τη θερμοκρασία (σε °C) και τη σχετική υγρασία (%) και να υπολογίζει τον [Heat Index (HI)](https://www.weather.gov/ama/heatindex), δηλαδή την αντιληπτή θερμοκρασία που συνδυάζει τις δύο τιμές. Επίσης να εμφανίζει το επίπεδο κινδύνου και σχετική περιγραφή, σύμφωνα με τον πίνακα ερμηνείας που βρίσκεται στο τέλος της εκφώνησης.

**Τρόπος υπολογισμού του Heat Index**

Η θερμοκρασία $T_C$ μετατρέπεται πρώτα σε βαθμούς Fahrenheit:

$T_F = 1.8 \times T_C + 32$

Στη συνέχεια ο Heat Index (σε °F) υπολογίζεται ως:

$HI_F = c1 + c2 \times T + c3 \times RH + c4 \times T_F \times RH + c5 \times T_F^2 + c6 \times RH^2 + c7 \times T_F^2\times RH + c8 \times T_F \times RH^2 + c9 \times T_F^2 \times RH^2$

όπου RH είναι η σχετική υγρασία και οι συντελεστές είναι:

| Συντελεστής | Τιμή |
|-------------|------|
| c1 | -42.379 |
| c2 | 2.04901523 |
| c3 | 10.14333127 |
| c4 | -0.22475541 |
| c5 | -0.00683783 |
| c6 | -0.05481717 |
| c7 | 0.00122874 |
| c8 | 0.00085282 |
| c9 | -0.00000199 |

Τέλος, η τιμή του Heat Index σε °C υπολογίζεται με:

$HI_C = \frac{HI_F - 32}{1.8}$

**Πίνακας ερμηνείας Heat Index**

| Heat Index (°C) | Επίπεδο κινδύνου | Περιγραφή |
|-----------------|------------------|------------|
| < 27             | Κανονικός        | Άνετος και ευχάριστος καιρός |
| 27 – 32          | Προσοχή          | Ελαφρώς δυσάρεστος, πιθανή κόπωση |
| 32 – 41          | Προειδοποίηση    | Πολύ ζεστός καιρός, χρειάζεται προσοχή |
| 41 – 54          | Έντονος κίνδυνος | Κίνδυνος θερμικής εξάντλησης |
| > 54             | Ακραίος κίνδυνος | Πολύ επικίνδυνη ζέστη, πιθανή θερμοπληξία |

```
Temperature = 30.0°C, RH = 70.0%
Heat Index = 35.0°C
```
