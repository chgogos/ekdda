1.1 Γράψτε κώδικα που να υπολογίζει τη θέση ενός σώματος που κινείται ευθύγραμμα και ομαλά. Να αναθέτει σε μεταβλητές την αρχική θέση σε μέτρα, την ταχύτητα σε χιλιόμετρα ανά ώρα, και τον χρόνο κίνησης σε λεπτά  στις τιμές 10 μέτρα, 60 χιλιόμετρα/ώρα και 30 λεπτά αντίστοιχα. Με βάση τον τύπο:

$x = x_0 + v*t$

να υπολογίζει και να εμφανίζει τη θέση x σε μέτρα με κατάλληλο μήνυμα, λαμβάνοντας υπόψη ότι στον τύπο, $x_0$ είναι αρχική θέση σε μέτρα, $v$ είναι η ταχύτητα σε μέτρα/δευτερόλεπτο και $t$ ο χρόνος σε δευτερόλεπτα.

```python
# Λύση 1.1
x0 = 10            # αρχική θέση σε μέτρα
v_kmh = 60         # ταχύτητα σε km/h
t_min = 30         # χρόνος σε λεπτά

v = v_kmh * 1000 / 3600   # ταχύτητα σε m/s
t = t_min * 60            # χρόνος σε s

x = x0 + v * t

print(f"Η τελική θέση του σώματος είναι {x:.1f} μέτρα")
```

1.2 Το καθαρό νερό σε κανονική ατμοσφαιρική πίεση έχει σημείο τήξης 0°C και σημείο βρασμού 100°C. Γράψτε κώδικα που να δέχεται από τον χρήστη τη θερμοκρασία T σε βαθμούς Φαρενάιτ και να εμφανίζει την κατάσταση του νερού:
- «Στερεή κατάσταση (πάγος)», αν T < 0°C
- «Υγρή κατάσταση (νερό)», αν 0 ≤ T < 100°C
- «Αέρια κατάσταση (υδρατμός)», αν T ≥ 100°C

Σημείωση: Η μετατροπή από βαθμούς Φαρενάιτ (°F) σε βαθμούς Κελσίου (°C) γίνεται με τον τύπο °C=(°F-32)*5/9.

```python
# Λύση 1.2
T_f = float(input("Δώσε τη θερμοκρασία σε βαθμούς Φαρενάιτ (°F): "))
T_c = (T_f - 32) * 5 / 9
print(f"Τ={T_f}°F, {T_c}°C")


if T_c < 0:
    print("Στερεή κατάσταση (πάγος)")
elif T_c < 100:
    print("Υγρή κατάσταση (νερό)")
else:
    print("Αέρια κατάσταση (υδρατμός)")
```

1.3  Γράψτε κώδικα που να δέχεται 2 βαθμούς από 1 μέχρι και 100. Αν απέχουν πάνω από 10 μονάδες τότε να δέχεται και τρίτο βαθμό. Σε κάθε περίπτωση να υπολογίζει τον τελικό βαθμό ως μέσο όρο των δύο υψηλότερων βαθμών και να τον εμφανίζει.

```python
# Λύση 1.3
b1 = int(input("Δώσε τον πρώτο βαθμό (1–100): "))
b2 = int(input("Δώσε τον δεύτερο βαθμό (1–100): "))

if abs(b1 - b2) > 10:
    b3 = int(input("Δώσε τον τρίτο βαθμό (1–100): "))
    print(f"Πρώτος βαθμός={b1}, δεύτερος βαθμός={b2}, τρίτος βαθμός={b3}")
    if b1 <= b2 and b1 <= b3:
        final = (b2 + b3) / 2
    elif b2 <= b1 and b2 <= b3:
        final = (b1 + b3) / 2
    else:
        final = (b1 + b2) / 2
else:
    print(f"Πρώτος βαθμός={b1}, δεύτερος βαθμός={b2}")
    final = (b1 + b2) / 2

print("Ο τελικός βαθμός είναι:", final)
```

1.4 Γράψτε κώδικα που να υλοποιεί τη [διαδικασία Kaprekar](https://en.wikipedia.org/wiki/Kaprekar%27s_routine). Δηλαδή να δέχεται έναν τετραψήφιο αριθμό να διατάσσει τα ψηφία του σε αύξουσα σειρά και σε φθίνουσα σειρά, να αφαιρεί το μεγαλύτερο από το μικρότερο και να επαναλαμβάνει τη διαδικασία με το αποτέλεσμα, μέχρι να μην προκύπτει αλλαγή στο αποτέλεσμα.

```python
# Λύση 1.4
n = int(input("Δώσε έναν τετραψήφιο αριθμό: "))

previous = -1
while n != previous:
    previous = n
    digits = f"{n:04d}"
    asc = "".join(sorted(digits))
    desc = "".join(sorted(digits, reverse=True))
    small = int(asc)
    large = int(desc)
    n = large - small
    print(f"{large:04d} - {small:04d} = {n:04d}")
print(f"Τελικό αποτέλεσμα: {n:04d}")
```

1.5 Γράψτε κώδικα που να υπολογίζει προσεγγιστικά τις ρίζες της εξίσωσης $f(x)=x^2-4$. Να εξετάζει όλες τις τιμές της μεταβλητής $x$ στο διάστημα $[-5,5]$ με βήμα 0.1, να υπολογίζει την τιμή της συνάρτησης $f(x)$ και να εντοπίζει τις τιμές του $x$ που η $f(x)$ βρίσκεται κοντά στο 0 (π.χ. η $f(x)$ είναι μικρότερη από $10^{-6}$).

```python
# Λύση 1.5
step = 0.1
x = -5.0

while x <= 5.0:   
    f = x**2 - 4
    if abs(f) <= 1e-6:
        print(f"Προσεγγιστική ρίζα: x = {x}")
    x += step
```
