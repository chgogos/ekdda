1.2 Να γραφεί πρόγραμμα σε Python που να υπολογίζει το εμβαδόν και την περίμετρο ενός τριγώνου με δεδομένες τις πλευρές του a, b, c. Το πρόγραμμα θα ζητά από τον χρήστη να εισάγει τις τρεις πλευρές και θα ελέγχει αν μπορούν να σχηματίσουν τρίγωνο (κάθε πλευρά μικρότερη από το άθροισμα των άλλων δύο). Αν οι τιμές που δόθηκαν δεν σχηματίζουν τρίγωνο, να εμφανίζεται κατάλληλο μήνυμα. Αν σχηματίζουν τρίγωνο, να υπολογίζεται η περίμετρος και το εμβαδόν του τριγώνου. Για το εμβαδόν $E$ μπορεί να χρησιμοποιηθεί ο τύπος του Ήρωνα:

$E = \sqrt{p*(p-a)(p-b)(p-c)}$, όπου $p = \frac{a+b+c}{2}$


````python
# Λύση 1.2
import math

a = float(input("Δώσε την πλευρά a: "))
b = float(input("Δώσε την πλευρά b: "))
c = float(input("Δώσε την πλευρά c: "))

if a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0:
    perimeter = a + b + c
    p = perimeter / 2
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))

    print(f"Το τρίγωνο {a=}, {b=}, {c=} είναι έγκυρο.")
    print(f"Περίμετρος: {perimeter}")
    print(f"Εμβαδόν: {area}")
else:
    print(f"Τα δεδομένα {a=}, {b=}, {c=} δεν σχηματίζουν τρίγωνο.")
````

1.3 Το πλήθος των τρόπων που μπορούν να επιλεγούν $k$ άτομα από $n$ άτομα δίνεται από τον τύπο $\binom{n}{k}=\frac{n!}{k!(n-k)!}$, όπου $n! = 1*2*3*...*n$.
Εμφανίστε το πληθος των διαφορετικών ομάδων που μπορούν να σχηματιστούν από 50 άτομα για μέγεθος ομάδα από 1 έως και 50

Δείτε τις συναρτήσεις `math.factorial()` και `math.comb()` από το package `math` της Python.

````python
# Λύση 1.3
import math

n = 50

print(
    f"Αριθμός διαφορετικών ομάδων από 1 έως {n} άτομα (υπολογισμός με τύπο παραγοντικών):"
)
for k in range(1, n + 1):
    c = math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
    print(f"{k} άτομα: {c}")
````

````python
# Λύση 1.3b
import math

n = 50

print(f"Αριθμός διαφορετικών ομάδων από 1 έως {n} άτομα:")
for k in range(1, n + 1):
    c = math.comb(n, k)  # Υπολογισμός συνδυασμού n επιλέγουμε k
    print(f"{k} άτομα: {c}")
````

1.4 Γράψτε πρόγραμμα σε Python που να βρίσκει όλα τα ζεύγη αριθμών $(i,j)$ με $1 \leq i \leq 20$ και $1 \leq j \leq 20$ για τα οποία ισχύει ότι το γινόμενο τους $i*j$ είναι διψήφιος αριθμός και άρτιος. Επίσης να εμφανίζει και το πλήθος των ζευγών για τα οποία ισχύει το παραπάνω. Το πρόγραμμα πρέπει να χρησιμοποιεί διπλή επανάληψη for ή while, και να εμφανίζει τα ζεύγη και το γινόμενο τους σε μορφή: 

```
i = 2, j = 6, i*j = 12
i = 3, j = 4, i*j = 12
...
```

````python
# Λύση 1.4
c = 0

for i in range(1, 21):
    for j in range(1, 21):
        product = i * j
        if 10 <= product <= 99 and product % 2 == 0:
            print(f"i = {i}, j = {j}, i*j = {product}")
            c += 1

print(f"Συνολικό πλήθος ζευγών: {c}")
````

1.5 Γράψτε κώδικα που να επιλέγει τυχαία αριθμούς με το `random.uniform` μέχρι το άθροισμά τους να γίνει μεγαλύτερο από 1. 

Επαναλάβατε το πείραμα 1000000 φορές. Κατά μέσο όρο πόσες φορές χρειάστηκε να επιλεγούν τιμές για να ξεπεράσει το άθροισμα την τιμή 1. Συγκρίνατε τον αριθμό αυτό με το `math.e`. 

Υπάρχει κάποιο όνομα για αυτό το πείραμα;

````python
# Λύση 1.5

import random

experiments = 1_000_000
count = 0
total1 = 0

for _ in range(experiments):
    total = 0
    count = 0
    while total <= 1:
        total += random.uniform(0, 1)
        count += 1
    total1 += count

average_count = total1 / experiments
print(f"Μέσος αριθμός επιλογών: {average_count:.4f}")
print(f"Σύγκριση με e ≈ {2.71828}")
````
