### Μεταβλητές και αριθμητικοί τελεστές
1.1 Υπολογίστε μια προσέγγιση του αριθμού $\pi$, κάνοντας την πράξη $4\left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - \frac{1}{11}\right)$, αναθέστε το αποτέλεσμα στη μεταβλητή `pi`. Ποια η διαφορά της τιμής που υπολογίσατε από την αποθηκευμένη σταθερά για το $\pi$ στο module `math`.

````python
# Λύση 1.1
import math

pi = 4 * (1 - 1/3 + 1/5 - 1/7 + 1/9 - 1/11)
difference = abs(math.pi - pi)
print("Προσέγγιση:", pi)
print("Διαφορά:", difference)
````

1.2 Γράψτε κώδικα που να δέχεται ένα ποσό που ένας υποθετικός πελάτης τράπεζας θέλει να κάνει ανάληψη από τον τραπεζικό του λογαριασμό. Το υπόλοιπο του λογαριασμού έστω ότι είναι ένας τυχαίος ακέραιος από 100 μέχρι και 1000 ευρώ (χρησιμοποιήστε την `random.randint(100,1000)`). Το πρόγραμμα να "επιτρέπει" την ανάληψη αν ο χρήστης εισάγει ποσό που είναι πολλαπλάσιο του 10 και δεν ξεπερνά τα 500 ευρώ και βέβαια δεν ξεπερνά επίσης το υπόλοιπο του λογαριασμού. Να εμφανίζει κατάλληλα μηνύματα για κάθε περίπτωση.

````python
# Λύση 1.2
import random

balance = random.randint(100, 1000)
print(f"Υπόλοιπο λογαριασμού: {balance}€")

amount = int(input("Ποσό ανάληψης: "))

if amount % 10 != 0:
    print("Σφάλμα: το ποσό πρέπει να είναι πολλαπλάσιο του 10.")
elif amount > 500:
    print("Σφάλμα: το ποσό δεν μπορεί να ξεπερνά τα 500€.")
elif amount > balance:
    print("Σφάλμα: ανεπαρκές υπόλοιπο λογαριασμού.")
else:
    balance -= amount
    print(f"Ανάληψη {amount}€ ολοκληρώθηκε με επιτυχία.")
    print(f"Νέο υπόλοιπο: {balance}€")
````

1.3  Γράψτε κώδικα που να προσομοιώνει 10 ρίψεις ενός ζαριού, χρησιμοποιώντας την `random.randint(1, 6)` για κάθε ρίψη. Εκτυπώστε το αποτέλεσμα κάθε ρίψης και στο τέλος, εμφανίστε πόσες φορές έπεσε 6.

````python
# Λύση 1.3
import random

count_sixes = 0
for i in range(10):
    roll = random.randint(1, 6)
    print(f"Ρίψη {i+1}: {roll}")
    if roll == 6:
        count_sixes += 1

print(f"Το 6 έπεσε {count_sixes} φορές.")
````

1.4 Γράψτε κώδικα που για το ακόλουθο κείμενο που βρίσκεται αποθηκευμένο στη μεταβλητή `text` χρησιμοποιώντας βρόχο for να εξετάζει κάθε λέξη του κειμένου, εμφανίζοντας για κάθε λέξη μεγαλύτερη από 3 χαρακτήρες που ξεκινά ή τελειώνει με φωνήεν το κείμενο της λέξης και τον αριθμό χαρακτήρων της. Στο τέλος να εμφανίζει πόσες λέξεις ξεκινούν με φωνήεν και πόσες τελειώνουν με φωνήεν.

```py
text = """Η Python ενδείκνυται ως γλώσσα εισαγωγική στον προγραμματισμό και στην επιστήμη των υπολογιστών γενικότερα με δεδομένη την απλότητα στη σύνταξή της 

Διεθνώς πολλά πανεπιστήμια αλλά και η δευτεροβάθμια εκπαίδευση σε πολλές χώρες υιοθετούν μια python first προσέγγιση εισαγωγής στον προγραμματισμό"""
```

````python
# Λύση 1.4
text = """Η Python ενδείκνυται ως γλώσσα εισαγωγική στον προγραμματισμό και στην επιστήμη των υπολογιστών γενικότερα με δεδομένη την απλότητα στη σύνταξή της 

Διεθνώς πολλά πανεπιστήμια αλλά και η δευτεροβάθμια εκπαίδευση σε πολλές χώρες υιοθετούν μια python first προσέγγιση εισαγωγής στον προγραμματισμό"""
vowels = "αειουηωάέίόύήώ"

start_vowel_count = 0
end_vowel_count = 0
words = text.split()

for word in words:
    if len(word) <= 3: 
        continue
    word = word.lower()
    starts_with_vowel = word[0] in vowels
    ends_with_vowel = word[-1] in vowels
    
    if starts_with_vowel:
        start_vowel_count += 1
    if ends_with_vowel:
        end_vowel_count += 1
    
    if starts_with_vowel or ends_with_vowel:
        print(f"Λέξη: {word} Μήκος: {len(word)}")

print(f"\nΣύνολο λέξεων που ξεκινούν με φωνήεν: {start_vowel_count}")
print(f"Σύνολο λέξεων που τελειώνουν με φωνήεν: {end_vowel_count}")

````

1.5 Χρησιμοποιώντας δύο εμφωλευμένους βρόχους δημιουργήστε 4 σειρές τυχαίων τιμών από το 1 μέχρι το 32 με 5 τιμές για κάθε σειρά. Για κάθε αριθμό να εμφανίζει την τιμή του μαζί με τη λέξη "Ζυγός" ή "Περιττός" ανάλογα με την τιμή του.

````python
# Λύση 1.5
import random

for i in range(4):  
    for j in range(5): 
        num = random.randint(1, 32)
        parity = "Ζυγός" if num % 2 == 0 else "Περιττός"
        print(f"{num:2} {parity}", end="\t")
    print() 
````
