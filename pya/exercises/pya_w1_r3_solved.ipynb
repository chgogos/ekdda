1.1 Αναθέστε σε μια μεταβλητή $x$ το αποτέλεσμα της πράξης $\frac{\sqrt{\pi}}{2^{0.9}}$ και σε μια μεταβλήτή $y$ το αποτέλεσμα της πράξης $x + \frac{e^x}{2}$. Υπολογίστε το αποτέλεσμα της έκφρασης $1 + 2^y + 3^y + 4^y$ εκτυπώστε ξεχωριστά το ακέραιο και τα 5 πρώτα ψηφία του δεκαδικού μέρος του αποτελέσματος. Λάβετε τις τιμές του $\pi$ και του $e$ από το module `math`.



````python
# Λύση 1.1
import math

x = math.sqrt(math.pi) / 2**0.9
y = x + math.e**x/2
z = 1 + 2**y + 3**y + 4**y
integer_part = int(z)
decimal_part = z - integer_part
print("Ακέραιο μέρος:", integer_part)
print("Πέντε πρώτα ψηφία δεκαδικού μέρος:", str(decimal_part)[2:7])
````

### Εντολές επιλογής
1.2 Γράψτε πρόγραμμα που να ζητά από τον χρήστη τη θερμοκρασία (σε °C) και τη σχετική υγρασία (%) και να υπολογίζει τον [Heat Index (HI)](https://www.weather.gov/ama/heatindex), δηλαδή την αντιληπτή θερμοκρασία που συνδυάζει τις δύο τιμές. Επίσης να εμφανίζει το επίπεδο κινδύνου και σχετική περιγραφή, σύμφωνα με τον πίνακα ερμηνείας που βρίσκεται στο τέλος της εκφώνησης.

**Τρόπος υπολογισμού του Heat Index**

Η θερμοκρασία $T_C$ μετατρέπεται πρώτα σε βαθμούς Fahrenheit:

$T_F = 1.8 \times T_C + 32$

Στη συνέχεια ο Heat Index (σε °F) υπολογίζεται ως:

$HI_F = c1 + c2 \times T + c3 \times RH + c4 \times T_F \times RH + c5 \times T_F^2 + c6 \times RH^2 + c7 \times T_F^2\times RH + c8 \times T_F \times RH^2 + c9 \times T_F^2 \times RH^2$

όπου RH είναι η σχετική υγρασία και οι συντελεστές είναι:

| Συντελεστής | Τιμή |
|-------------|------|
| c1 | -42.379 |
| c2 | 2.04901523 |
| c3 | 10.14333127 |
| c4 | -0.22475541 |
| c5 | -0.00683783 |
| c6 | -0.05481717 |
| c7 | 0.00122874 |
| c8 | 0.00085282 |
| c9 | -0.00000199 |

Τέλος, η τιμή του Heat Index σε °C υπολογίζεται με:

$HI_C = \frac{HI_F - 32}{1.8}$

**Πίνακας ερμηνείας Heat Index**

| Heat Index (°C) | Επίπεδο κινδύνου | Περιγραφή |
|-----------------|------------------|------------|
| < 27             | Κανονικός        | Άνετος και ευχάριστος καιρός |
| 27 – 32          | Προσοχή          | Ελαφρώς δυσάρεστος, πιθανή κόπωση |
| 32 – 41          | Προειδοποίηση    | Πολύ ζεστός καιρός, χρειάζεται προσοχή |
| 41 – 54          | Έντονος κίνδυνος | Κίνδυνος θερμικής εξάντλησης |
| > 54             | Ακραίος κίνδυνος | Πολύ επικίνδυνη ζέστη, πιθανή θερμοπληξία |

Για παράδειγμα για τιμές εισόδου θερμοκρασίας 30°C και σχετικής υγρασίας 70% θα πρέπει να εμφανίζεται μήνυμα της μορφής:

```
Temperature = 30.0°C, RH = 70.0%
Heat Index = 35.0°C  
Προειδοποίηση: Πολύ ζεστός καιρός, χρειάζεται προσοχή.
```




````python
# Λύση 1.2
T_C = float(input("Δώσε τη θερμοκρασία σε °C: "))
RH = float(input("Δώσε τη σχετική υγρασία σε %: "))

T_F = 1.8 * T_C + 32

c1 = -42.379
c2 = 2.04901523
c3 = 10.14333127
c4 = -0.22475541
c5 = -0.00683783
c6 = -0.05481717
c7 = 0.00122874
c8 = 0.00085282
c9 = -0.00000199

HI_F = (
    c1
    + c2 * T_F
    + c3 * RH
    + c4 * T_F * RH
    + c5 * T_F**2
    + c6 * RH**2
    + c7 * T_F**2 * RH
    + c8 * T_F * RH**2
    + c9 * T_F**2 * RH**2
)

HI_C = (HI_F - 32) / 1.8

if HI_C < 27:
    label = "Κανονικός: Άνετος και ευχάριστος καιρός"
elif HI_C < 32:
    label = "Προσοχή: Ελαφρώς δυσάρεστος, πιθανή κόπωση"
elif HI_C < 41:
    label = "Προειδοποίηση: Πολύ ζεστός καιρός, χρειάζεται προσοχή"
elif HI_C < 54:
    label = "Έντονος κίνδυνος: Κίνδυνος θερμικής εξάντλησης"
else:
    label = "Ακραίος κίνδυνος: Πολύ επικίνδυνη ζέστη, πιθανή θερμοπληξία"

print(f"Temperature = {T_C:.1f}°C, RH = {RH:.1f}%")
print(f"Heat Index = {HI_C:.1f}°C")
print(label)
````

1.3  Γράψτε κώδικα που να υπολογίζει τον Μέγιστο Κοινό Διαιρέτη (ΜΚΔ) από δύο θετικούς ακεραίους αριθμούς που θα δίνει ο χρήστης. Ο υπολογισμός να γίνεται ως εξής:

Επαναλαμβάνετε τη διαδικασία διαίρεσης του μεγαλύτερου αριθμού με τον μικρότερο και αντικαθιστάτε τον μεγαλύτερο με το υπόλοιπο της διαίρεσης, μέχρι το υπόλοιπο να γίνει μηδέν. Όταν αυτό συμβεί, ο μικρότερος αριθμός της τελευταίας διαίρεσης είναι ο ΜΚΔ (αλγόριθμος του Ευκλείδη για την εύρεση ΜΚΔ).

Το αποτέλεσμα να εμφανίζεται ως εξής (π.χ. αν ο χρήστης εισάγει `144` και `160`):
```
Ο ΜΚΔ των 144 και 160 είναι: 16
```

````python
# Λύση 1.3

a = int(input("Δώσε τον πρώτο αριθμό: "))
b = int(input("Δώσε τον δεύτερο αριθμό: "))
ca, cb = a, b

if a < b:
    a, b = b, a

while b != 0:
    r = a % b
    a = b
    b = r

print(f"Ο ΜΚΔ των {ca} και {cb} είναι:", a)
````

1.4 Γράψτε κώδικα που να εμφανίζει όλους τους αριθμούς Harshad στο διάστημα `[100, 200]`. Ένας αριθμός είναι [αριθμός Harshad](https://en.wikipedia.org/wiki/Harshad_number) αν διαιρείται ακριβώς με το άθροισμα των ψηφίων του (π.χ. ο αριθμός `140` είναι αριθμός Harshad διότι διαιρείται ακριβώς με το `1 + 4 = 5`)


````python
# Λύση 1.4
for n in range(100, 201):
    digit_sum = 0
    for d in str(n):
        digit_sum += int(d)
    if n % digit_sum == 0:
        print(n)
````

1.5 Γράψτε πρόγραμμα που να εκτυπώνει τρία σχήματα με αστεράκια για έναν θετικό ακέραιο `n` που δίνεται από τον χρήστη: πρώτα ένα τετράγωνο `n x n`, μετά ένα τρίγωνο με αύξουσα σειρά αστεριών από `1` έως `n` και τέλος ένα τρίγωνο με φθίνουσα σειρά αστεριών από `n` έως `1`.  

**Παράδειγμα για n = 4:**  

```
Σχήμα 1: Τετράγωνο
****
****
****
****

Σχήμα 2: Τρίγωνο
*
**
***
****

Σχήμα 3: Αντίστροφο τρίγωνο
****
***
**
*
```

````python
# Λύση 1.5
n = int(input("Δώσε έναν θετικό ακέραιο n: "))
print("Για n =", n)

print("Σχήμα 1: Τετράγωνο")
for i in range(n):
    for j in range(n):
        print("*", end="")
    print()  

print("\nΣχήμα 2: Τρίγωνο")
for i in range(1, n+1):
    for j in range(i):
        print("*", end="")
    print()

print("\nΣχήμα 3: Αντίστροφο τρίγωνο")
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
````
