Τέλος να εμφανίζονται κάνοντας χρήση της `__str__()` οι μαθητές που έχουν προβιβάσιμο βαθμό. 


Δείτε ένα παράδειγμα εκτέλεσης που θα παράγει ο ζητούμενος κώδικας:
```cmd
Δώσε όνομα: Μαρία
Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): 18

Δώσε όνομα: Γιάννης
Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): 25
Σφάλμα: Ο βαθμός πρέπει να είναι είναι >=0 και <=20 ή να είναι -1
Παρακαλώ προσπάθησε ξανά.
Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): 9

Δώσε όνομα: Ελένη
Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): αβγ
Σφάλμα: invalid literal for int() with base 10: 'αβγ'
Παρακαλώ προσπάθησε ξανά.
Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): 12

Μαθητές με προβιβάσιμο βαθμό:
Όνομα='Μαρία', βαθμός=18
Όνομα='Ελένη', βαθμός=12
```

````python
# Λύση
class Student:
    def __init__(self, name):
        self.name = name
        self.grade = -1

    def set_grade(self, grade):
        if grade == -1 or (0 <= grade <= 20):
            self.grade = grade
        else:
            raise ValueError(
                "Ο βαθμός πρέπει να είναι είναι >=0 και <=20 ή να είναι -1"
            )

    def passed(self):
        return self.grade >= 10

    def __str__(self):
        return f"Όνομα='{self.name}', βαθμός={self.grade}"


students = []
for i in range(3):
    name = input("Δώσε όνομα: ")
    student = Student(name)
    while True:
        try:
            grade_input = input("Δώσε βαθμό (0-20 ή -1 αν δεν προσήλθε): ")
            grade = int(grade_input)
            student.set_grade(grade)
            break
        except ValueError as e:
            print("Σφάλμα:", e)
            print("Παρακαλώ προσπάθησε ξανά.")
    students.append(student)


print("Μαθητές με προβιβάσιμο βαθμό:")
for s in students:
    if s.passed():
        print(s)
````
