```python
# Λύση

class Recipe:
    def __init__(self, name, instructions):
        self.name = name
        self.instructions = instructions
        self.ingredients = []

    def add_ingredient(self, item):
        self.ingredients.append(item)

    def __str__(self):
        ingredients_str = "\n".join(f" - {i}" for i in self.ingredients)
        return (
            f"Συνταγή: {self.name}\n"
            f"Υλικά:\n{ingredients_str}\n"
            f"Οδηγίες: {self.instructions}\n"
        )


# Δημιουργία αντικειμένου για τη Χωριάτικη σαλάτα
recipe1 = Recipe(
    "Χωριάτικη σαλάτα",
    "Κόψτε τις ντομάτες, το αγγούρι και το κρεμμύδι σε κομμάτια. "
    "Προσθέστε τη φέτα και τις ελιές, περιχύστε με ελαιόλαδο και ανακατέψτε ελαφρά."
)

recipe1.add_ingredient("ντομάτες")
recipe1.add_ingredient("αγγούρι")
recipe1.add_ingredient("κρεμμύδι")
recipe1.add_ingredient("φέτα")
recipe1.add_ingredient("ελιές")
recipe1.add_ingredient("ελαιόλαδο")


# Δημιουργία αντικειμένου για τις Τηγανίτες
recipe2 = Recipe(
    "Τηγανίτες",
    "Ανακατέψτε το αλεύρι, το γάλα και τα αυγά σε ένα μπολ μέχρι να δημιουργηθεί "
    "ομοιογενές μείγμα. Ρίξτε μικρές ποσότητες σε ζεστό τηγάνι και ψήστε μέχρι "
    "να ροδίσουν και από τις δύο πλευρές."
)

recipe2.add_ingredient("αλεύρι")
recipe2.add_ingredient("γάλα")
recipe2.add_ingredient("αυγά")
recipe2.add_ingredient("ζάχαρη")
recipe2.add_ingredient("βούτυρο")


# Εμφάνιση των συνταγών
print(recipe1)
print(recipe2)
```

### Β. Βαθιά αντιγραφή αντικειμένων, κληρονομικότητα
Επεκτείνετε τον προηγούμενο κώδικα με τα ακόλουθα:
- Να οριστεί κατάλληλος τρόπος δημιουργίας αντιγράφου αντικειμένου της κλάσης Recipe έτσι ώστε να πραγματοποιείται βαθιά αντιγραφή της λίστας των υλικών, δηλαδή το νέο αντικείμενο να διαθέτει δική του ανεξάρτητη λίστα. Δημιουργήστε ένα αντίγραφο της συνταγής χωριάτικης σαλάτας και αφαιρέστε από το αντίγραφο το κρεμμύδι.
- Ορίστε την υποκλάση της Recipe με όνομα Dessert, η οποία να προσθέτει επιπλέον το γνώρισμα calories για την αποθήκευση των θερμίδων του γλυκού. Δημιουργήστε ένα νέο αντικείμενο με όνομα "σοκολάτα-μπανάνα" με τα ακόλουθα γνωρίσματα.

```python
# Λύση

class Recipe:
    def __init__(self, name, instructions):
        self.name = name
        self.instructions = instructions
        self.ingredients = []

    def add_ingredient(self, item):
        self.ingredients.append(item)

    def copy(self):
        new_recipe = Recipe(self.name, self.instructions)
        new_recipe.ingredients = self.ingredients.copy()  # βαθιά αντιγραφή λίστας
        return new_recipe

    def __str__(self):
        ingredients_str = "\n".join(f" - {ingredient}" for ingredient in self.ingredients)
        return (
            f"Συνταγή: {self.name}\n"
            f"Υλικά:\n{ingredients_str}\n"
            f"Οδηγίες: {self.instructions}\n"
        )


class Dessert(Recipe):
    def __init__(self, name, instructions, calories):
        super().__init__(name, instructions)
        self.calories = calories

    def __str__(self):
        ingredients_str = "\n".join(f" - {ingredient}" for ingredient in self.ingredients)
        return (
            f"Επιδόρπιο: {self.name}\n"
            f"Θερμίδες: {self.calories}\n"
            f"Υλικά:\n{ingredients_str}\n"
            f"Οδηγίες:\n{self.instructions}\n"
        )


# Συνταγή 1
recipe1 = Recipe(
    "Χωριάτικη σαλάτα",
    "Κόψτε τις ντομάτες, το αγγούρι και το κρεμμύδι σε κομμάτια. "
    "Προσθέστε τη φέτα και τις ελιές, περιχύστε με ελαιόλαδο και ανακατέψτε ελαφρά."
)

recipe1.add_ingredient("ντομάτες")
recipe1.add_ingredient("αγγούρι")
recipe1.add_ingredient("κρεμμύδι")
recipe1.add_ingredient("φέτα")
recipe1.add_ingredient("ελιές")
recipe1.add_ingredient("ελαιόλαδο")


# Συνταγή 2
recipe2 = Recipe(
    "Τηγανίτες",
    "Ανακατέψτε το αλεύρι, το γάλα και τα αυγά σε ένα μπολ μέχρι να δημιουργηθεί "
    "ομοιογενές μείγμα. Ψήστε σε τηγάνι μέχρι να ροδίσουν."
)

recipe2.add_ingredient("αλεύρι")
recipe2.add_ingredient("γάλα")
recipe2.add_ingredient("αυγά")
recipe2.add_ingredient("ζάχαρη")
recipe2.add_ingredient("βούτυρο")


# Δημιουργία αντιγράφου της χωριάτικης σαλάτας
recipe1_copy = recipe1.copy()

# Αφαίρεση κρεμμυδιού από το αντίγραφο
recipe1_copy.ingredients.remove("κρεμμύδι")


# Δημιουργία επιδορπίου
dessert = Dessert(
    "σοκολάτα-μπανάνα",
    "Λιώστε τη σοκολάτα και ανακατέψτε με κομμένες μπανάνες.",
    2500
)

dessert.add_ingredient("σοκολάτα")
dessert.add_ingredient("μπανάνες")


# Εμφάνιση αποτελεσμάτων
print("Αρχική συνταγή:")
print(recipe1)

print("Αντίγραφο χωρίς κρεμμύδι:")
print(recipe1_copy)

print("Επιδόρπιο:")
print(dessert)
```
